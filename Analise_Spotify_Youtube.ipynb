```python
import pandas as pd
```

```python
df = pd.read_csv('Spotify_Youtube.csv')
df.info()
```

```python
### primeiro passo remover essa coluba 'Unnamed: 0' que nao tem utilidade
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
### vamos todos os artistas em nosso dataset, quero ver todos os nomes dos artistas em um novo dataframe
df_artists = df[['Artist']].drop_duplicates()
df_artists
# quero contar quantos artistas diferentes temos no dataset
num_artists = df_artists.shape[0]
print(f'Temos {num_artists} artistas diferentes no dataset.')
```

```python
### vamos verifcar os artistas com mais musicas na lista
df['Artist'].value_counts().head(10)
```

```python
### vamos ver se tem alguma musica repetida
df['Title'].value_counts().head(10) 
```

```python
# vamos ver s√≥ as musicas do artista Gorillaz
df[df['Artist'] == 'Gorillaz']
```

```python
# vamos deletar as musicas repetidas, se houver
df = df.drop_duplicates(subset=['Title', 'Artist'])
df
```

```python
### vamos ver se existe alguma musica do Bruno Mars
df[df['Artist'] == 'Bruno Mars']
```

```python
### quero mostrar as 5 musicas com mais views no youtube
df.sort_values(by='Views', ascending=False).head(5)
```
